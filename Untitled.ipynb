### Téléchargement des donées sur consommation d'energie et les émissions de gaz à effet de serre

```python

paris=pd.read_csv("https://koumoul.com/data-fair/api/v1/datasets/dpe-75/data-files/dpe-75.csv")
Seine_et_Marne=pd.read_csv("https://koumoul.com/data-fair/api/v1/datasets/dpe-77/data-files/dpe-77.csv")
Yvelines=pd.read_csv("https://koumoul.com/data-fair/api/v1/datasets/dpe-78/data-files/dpe-78.csv")
Essonne=pd.read_csv("https://koumoul.com/data-fair/api/v1/datasets/dpe-91/data-files/dpe-91.csv")
Hauts_de_Seine=pd.read_csv("https://koumoul.com/data-fair/api/v1/datasets/dpe-92/data-files/dpe-92.csv")
Seine_saint_Denis=pd.read_csv("https://koumoul.com/data-fair/api/v1/datasets/dpe-93/data-files/dpe-93.csv")
Val_de_Marne=pd.read_csv("https://koumoul.com/data-fair/api/v1/datasets/dpe-94/data-files/dpe-94.csv")
Val_Oise=pd.read_csv("https://koumoul.com/data-fair/api/v1/datasets/dpe-95/data-files/dpe-95.csv")
```

```python
data=pd.concat([paris,Seine_et_Marne,Yvelines,Essonne,Hauts_de_Seine,Seine_saint_Denis,Val_de_Marne,Val_Oise])
```

# II. Quelques statisques descriptives

### Consommation  et emission de gaz d'effet de serre totale par departement

```python
for i in ['consommation_energie','estimation_ges']:
    departement=data.groupby("tv016_departement_departement")[i].sum()
    departement=departement.sort_values()
    plt.barh(departement.index,departement.values,height=0.8,alpha=0.5,color='black',left=0.1)
    plt.title(f"{i} totale par Partement")
    plt.show()
```

```python
for i in ['consommation_energie','estimation_ges']:
    departement=data.groupby("tv016_departement_departement")[i].mean()
    departement=departement.sort_values()
    plt.barh(departement.index,departement.values,height=0.8,alpha=0.5,color='black',left=1)
    plt.title(f"{i} moyenne par Partement")
    plt.show()
```

### Les Communes les plus énergivores et qui émettent les plus de gaz a effet de serre 

```python
for i in ['consommation_energie','estimation_ges']:
    ccommune=data.groupby("commune")[i].sum()
    ccommune=ccommune.sort_values().head(15)
    plt.barh(ccommune.index,ccommune.values,height=0.8,alpha=0.5,color='black',left=0.1)
    plt.title(f"{i} totale par commune")
    plt.show()
```

```python
for i in ['consommation_energie','estimation_ges']:
    ccommune=data.groupby("commune")[i].mean()
    ccommune=ccommune.sort_values().head(15)
    plt.barh(ccommune.index,ccommune.values,height=0.8,alpha=0.5,color='black',left=0.1)
    plt.title(f"{i} moyenne par commune")
    plt.show()
```
